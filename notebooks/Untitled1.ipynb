```python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv('final_cleaned_twitter_bot_data.csv')

print(f"Original shape: {df.shape}")
```

```python
# Create follower-to-friend ratio
df['follower_friend_ratio'] = df['followers_count'] / (df['friends_count'] + 1)

print("âœ“ Created: follower_friend_ratio")
```

```python
# Create tweets-per-follower ratio
df['tweets_per_follower'] = df['statuses_count'] / (df['followers_count'] + 1)

print("âœ“ Created: tweets_per_follower")
```

```python
# Create profile completeness score
df['profile_completeness'] = (
    ((df['description'].str.len() > 0).astype(int)) +
    (~df['default_profile']).astype(int) +
    (~df['default_profile_image']).astype(int) +
    df['geo_enabled'].astype(int)
) / 4

print("âœ“ Created: profile_completeness")
```

```python
# Check if description contains URL (http or www)
df['description_has_url'] = df['description'].str.contains(
    'http|www', 
    case=False, 
    na=False
).astype(int)

print("âœ“ Created: description_has_url")
```

```python
# Check if username contains numbers (e.g., bot123456)
df['name_has_numbers'] = df['screen_name'].str.contains(
    r'\d', 
    na=False, 
    regex=True
).astype(int)

print("âœ“ Created: name_has_numbers")
```

```python
# Save to NEW CSV
new_filename = 'twitter_bot_final_features.csv'
df.to_csv(new_filename, index=False)
```

```python
import pandas as pd

# Load the file
df = pd.read_csv('twitter_bot_final_features.csv')

print("="*70)
print("ðŸ” MISSING VALUES BREAKDOWN")
print("="*70)

# Check missing values per column
missing_info = df.isnull().sum()
missing_info = missing_info[missing_info > 0].sort_values(ascending=False)

if len(missing_info) > 0:
    print("\nColumns with missing values:\n")
    for col, count in missing_info.items():
        percentage = (count / len(df)) * 100
        print(f"  {col:30s} {count:>6,} ({percentage:>5.2f}%)")
    
    print("\n" + "="*70)
    print("EXPLANATION")
    print("="*70)
    
    # Check if it's the description column
    if 'description' in missing_info.index:
        print("\nâœ… Expected: 'description' has missing values")
        print("   â†’ This is NORMAL - many Twitter accounts have empty bios")
        print("   â†’ Already handled during cleaning (empty strings exist)")
        print("   â†’ Safe to keep as-is for modeling")
    
    print("\n" + "="*70)
    print("RECOMMENDATION")
    print("="*70)
    print("\nOption 1: Keep as-is (RECOMMENDED)")
    print("  â€¢ Empty descriptions are a valid feature")
    print("  â€¢ Bots often have no bio (useful signal!)")
    print("  â€¢ 'has_description' feature already captures this")
    
    print("\nOption 2: Fill missing descriptions")
    print("  â€¢ Replace NaN with empty string ''")
    print("  â€¢ Won't change model performance")
    
else:
    print("âœ… No missing values found!")

print("\n" + "="*70)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('twitter_bot_final_features.csv')

# --- 2. Histogram for Average Tweets Per Day ---
plt.figure(figsize=(10, 6))

plt.hist(df['average_tweets_per_day'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Average Tweets Per Day')
plt.xlabel('Average Tweets Per Day')
plt.ylabel('Frequency (Log Scale)')
plt.yscale('log') 
plt.grid(axis='y', alpha=0.75)
plt.savefig('histogram_avg_tweets_per_day.png')
plt.close()

# --- 3. Histogram for Profile Completeness ---
plt.figure(figsize=(10, 6))
plt.hist(df['profile_completeness'], bins=20, color='lightcoral', edgecolor='black')
plt.title('Distribution of Profile Completeness')
plt.xlabel('Profile Completeness (0.0 to 1.0)')
plt.ylabel('Frequency (Count of Users)')
plt.grid(axis='y', alpha=0.75)
plt.savefig('histogram_profile_completeness.png')
plt.close()

print("Histograms saved as: histogram_avg_tweets_per_day.png and histogram_profile_completeness.png")
```

```python
plt.figure(figsize=(10, 6))
order = [0.0, 0.25, 0.50, 0.75, 1.0]
sns.countplot(
    data=df,
    x='profile_completeness',
    hue='account_type',
    order=order,
    palette=['lightcoral', 'skyblue'],
    edgecolor='black'
)

plt.title('Profile Completeness by Account Type (Human vs. Bot)')
plt.xlabel('Profile Completeness Score')
plt.ylabel('Count of Users (Frequency)')
plt.legend(title='Account Type')
plt.savefig('profile_completeness_by_account_type.png')
plt.close()
```

```python
plt.figure(figsize=(7,5))
sns.histplot(
    data=df,
    x="friends_count",
    hue="account_type",
    bins=50,
    element="step",
    log_scale=True,
    stat="density",
    common_norm=False
)
plt.title("Distribution of Friends Count by Account Type")
plt.xlabel("Friends Count (log scale)")
plt.ylabel("Density")
plt.show()
```

```python
sns.boxplot(data=df, x="account_type", y="account_age_days")
plt.title("Account Age by Account Type")
plt.show()
```
